## left join

```python
summer = pd.read_csv('summer.csv')
dictionary = pd.read_csv('dictionary.csv')
```

```python
summer.head()
```

```python
dictionary.head()
```

```python
len(summer),len(dictionary)
```

```python
dictionary.set_index('Code',drop=True, inplace=True)
```

```python
summer_new = summer.merge(dictionary, how='left', left_on='Country', right_index=True, indicator=True)
```

```python
summer_new.head()
```

```python
mapper={'Country_x':'country_code', 'Country_y':'country'}
summer_new.rename(columns=mapper, inplace=True)
```

```python
summer_new.loc[summer_new.country.isnull(),'country_code'].value_counts(dropna=False)
```

```python
summer_new.info()
```
